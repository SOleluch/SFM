```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Fréquences
f1 = 10
f2 = 20
f3 = 40

#Dimension, Fe >= 2*max(f)
dim = 80

#Nombre de données par classe
n = 2000

#Points d'évaluation
t = np.linspace(1, 1000, n)

#Bruits
noise = np.random.normal(0, 1, t.shape)

#Sinus
sinus1 = np.sin(2*np.pi*f1*t)
sinus2 = np.sin(2*np.pi*f2*t)
sinus3 = np.sin(2*np.pi*f3*t)

#Sinus avec bruits
sinus1Bruit = noise + np.sin(2*np.pi*f1*t)
sinus2Bruit = noise + np.sin(2*np.pi*f2*t)
sinus3Bruit = noise + np.sin(2*np.pi*f3*t)
```

```python
plt.subplot(3,1,1)
plt.plot(t, sinus1)
plt.title('Sinus1')
plt.show()

plt.subplot(3,1,2)
plt.plot(t, sinus2)
plt.title('Sinus2')
plt.show()

plt.subplot(3,1,3)
plt.plot(t, sinus3)
plt.title('Sinus3')
plt.show()
```

```python
plt.subplot(3,1,1)
plt.plot(t, sinus1Bruit)
plt.title('Sinus1')
plt.show()

plt.subplot(3,1,2)
plt.plot(t, sinus2Bruit)
plt.title('Sinus2')
plt.show()

plt.subplot(3,1,3)
plt.plot(t, sinus3Bruit)
plt.title('Sinus3')
plt.show()
```

```python
df_sinus1 = pd.DataFrame(sinus1, columns=['sinus'])
df_sinus2 = pd.DataFrame(sinus2, columns=['sinus'])
df_sinus3 = pd.DataFrame(sinus3, columns=['sinus'])

df_sinus1Bruit = pd.DataFrame(sinus1Bruit, columns=['sinus'])
df_sinus2Bruit = pd.DataFrame(sinus2Bruit, columns=['sinus'])
df_sinus3Bruit = pd.DataFrame(sinus3Bruit, columns=['sinus'])

df_sinus1.to_csv('sinus1.csv')
df_sinus2.to_csv('sinus2.csv')
df_sinus3.to_csv('sinus3.csv')

df_sinus1Bruit.to_csv('sinus1Bruit.csv')
df_sinus2Bruit.to_csv('sinus2Bruit.csv')
df_sinus3Bruit.to_csv('sinus3Bruit.csv')
```
